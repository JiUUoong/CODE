```python
# 표현 학습
# 유사도가 높으면 벡터공간에서 가깝게
# observed(observable; 관찰 가능한) space <--> hidden(latent; 은닉) space
```

```python
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
```

```python
mnist = fetch_openml("mnist_784", as_frame=False)
```

```python
X = mnist.data
y = mnist.target
X.shape
```

```python
pca = PCA(3)
X3d = pca.fit_transform(X)
X3d.shape
print(sum(pca.explained_variance_ratio_))
```

```python
import plotly.express as px
import pandas as pd

df = pd.DataFrame({"PC0": X3d[:, 0],
                   "PC1": X3d[:, 1],
                   "PC2": X3d[:, 2]})
column_names = df.columns
fig = px.scatter_3d(df, x=column_names[0], y=column_names[1], z=column_names[2], color=y)
fig.update_traces(marker_size=0.5)
fig.show()
```
